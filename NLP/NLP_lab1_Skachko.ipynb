```python
words = [word.lower() for word in text8 if word.isalpha()]

fdist = nltk.FreqDist(words)

hapaxes = fdist.hapaxes()
```

```python
hapaxes
```

```python
less_than_5 = [word for word, freq in fdist.items() if freq < 5]
less_than_5
```

```python
word_lengths = [len(word) for word in words]
word_freqs = [fdist[word] for word in words]

plt.figure(figsize=(10, 6))
plt.scatter(word_lengths, word_freqs, alpha=0.5)
plt.title('Mapping from Word Length to Word Frequency')
plt.xlabel('Word Length')
plt.ylabel('Word Frequency')
plt.grid(True)
plt.show()
```

```python

from nltk.tokenize import sent_tokenize, word_tokenize
from collections import Counter

# Розділяємо текст на речення
sentences = sent_tokenize(" ".join(text8.tokens))

last_words = []

# Проходимося по кожному реченню
for sentence in sentences:
    words = word_tokenize(sentence)
    # Видаляємо знаки пунктуації з кінця речення 
    while words[-1] in [".", "?", "!"]:
        words.pop()
   
    last_words.append(words[-1])

# Обчислюємо найбільш часто зустрічаємі слова в кінці речень
most_common_last_words = Counter(last_words).most_common(10)

print("10 найбільш часто зустрічаємі слова в кінці речень:")
for word, count in most_common_last_words:
    print(word, "-", count)
```

```python
import re

def sum_monetary_amounts(tokens):
    # Регулярний вираз для виявлення грошових сум
    pattern = r'\$\d+\.\d{1,2}'

    # Знаходимо всі входження грошових сум у тексті
    matches = re.findall(pattern, ' '.join(tokens))

    # Конвертуємо входження у числа з плаваючою точкою
    amounts = [float(match[1:]) for match in matches]

    total = sum(amounts)
    return total


with open('$.txt', 'r') as file:
    text = file.read()
    tokens = text.split()

total_amount = sum_monetary_amounts(tokens)

print("Загальна сума грошових виплат:", total_amount)
```

```python
import json
import re

# Створення початкового JSON файлу з заданими даними
initial_data = {
    "HTTPHost": "localhost",
    "HTTPPort": 8545,
    "Config": {
        "Enabled": True,
        "Server": "wakuv2.prod",
        "Nodes": [
            "8.210.222.231",
            "168.166.125.145",
            "192.168.0.1",
            "8.210.222.234",
            "8.210.222.588"
        ]
    }
}

# Запис початкового JSON файлу
with open('initial_data.json', 'w') as file:
    json.dump(initial_data, file, indent=4)

with open('initial_data.json', 'r') as file:
    data = json.load(file)

data['Config']['Nodes'] = [node for node in data['Config']['Nodes'] if not re.match(r'^8\.', node)]

server_name, server_type = re.match(r'(\w+)\.(\w+)', data['Config']['Server']).groups()
data['Config']['Server'] = {"name": server_name, "type": server_type}

# Запис відредагованого JSON файлу
with open('edited_data.json', 'w') as file:
    json.dump(data, file, indent=4)
```
